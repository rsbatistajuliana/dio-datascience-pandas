```python
#veio estranho, o separador é ;
#como mudar o separador? parametro da leitura sep=";" - muda o separador , para ;
df = pd.read_csv('/content/drive/MyDrive/desafio-projeto-2/datasets/Gapminder.csv',error_bad_lines=False,sep=";")
```

```python
df.head()
```

```python
#mudar nome das colunas e troca em df
df= df.rename(columns={"country":"Pais", 
                   "continent": "Continente",
                   "year":"Ano",
                   "lifeExp":"Expectativa de vida",
                   "pop":"Pop. Total",
                   "gdpPercap":"PIBpercapta" })
df.head(10)
```

```python
#describe retorna infos estatisticas
df.describe()
#STD - DESVIO PADRÃO
#quartis - 25%, 50% - mediana,75%
```

```python
#fazer filtro no df
#quero dados de um continente
df["Continente"].unique()
```

```python
#quero retornar apenas valores de oceania - FILTROS
Oceania = df.loc[df["Continente"]== "Oceania"]
Oceania.head()
```

```python
#Agrupamento de dados - útil para SQL
#Agrupar por continente mas quero trazer quantos paises por continente
df.groupby("Continente")["Pais"].nunique() #contagem distinta
```

```python
#Para cada ano qual é a expectativa de vida média?
df.groupby("Ano")["Expectativa de vida"].mean()
```

```python
#Media de uma coluna
df["PIBpercapta"].mean()
```
